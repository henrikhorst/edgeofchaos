```python
import random
from random import randrange
random.seed(42)
edges=[]

for j in range(len(list(G.edges()))):
    max_value=0
    for key, value in node_dict.items():
        if len(value) > max_value:
            max_value = len(value)
            max_key = key
    edge = node_dict[max_key][randrange(max_value)]
    #print(edge)
    edges.append(edge)
    #print(node_dict[max_key])
    node_dict[max_key].remove(edge)
    #print(node_dict[max_key])
    #print(node_dict[str(edge[1]+1)])
    node_dict[str(edge[1]+1)].remove((edge[1],edge[0]))
    #print(node_dict[str(edge[1]+1)])
    
```

### Oscillators

```python
def create_oscillator(Rin_a= '40k', Ca_gnd= '20u', Rc_gnd= '200', Rin_g= '10k', Rg_gnd= '20k',
                 input_connection= 'V1', output_connection= {}):
    '''
    This function creates a standard oscillator
    '''
    return {'Rin_a': Rin_a, 'Ca_gnd': Ca_gnd, 'Rc_gnd': Rc_gnd, 'Rin_g': Rin_g, 'Rg_gnd': Rg_gnd,
                 'input_connection': input_connection, 'output_connection': output_connection}
```

```python
num_Os=32 # how many oscillators
oscillators = []
oscillator_config=[]
# now manually create the desired component values below
Rin_as=[*list(range(40,48)), *list(np.arange(40.5,48.5,1)), *list(np.arange(40.25,48.25,1)), *list(np.arange(40.375,48.375,1))]
Ca_gnds= 32*[20]
input_connections = [f'V{x}' for x in range(1,num_Vs+1)]
output_connections = [dict() for i in range(num_Os)]

for edge in edges:
    output_connections[edge[0]][str(edge[1]+1)]= ['excitatory',{'R_coupling':'47k', 'C_coupling': '1u'}]
    
#output_connections[0]= {'2': ['excitatory',{'R_coupling':'94k', 'C_coupling': '2u'}]}
#output_connections[1] = {'3': ['inhibitory',{'R_coupling':'1'}]}
for i in range(num_Os):
    oscillator_config.append({'Rin_a': f'{Rin_as[i]}k', 'Ca_gnd': f'{Ca_gnds[i]}u', 'input_connection': input_connections[i], 'output_connection': output_connections[i]})
```

```python
for i in range(num_Os):
    oscillator= create_oscillator(**oscillator_config[i])
    oscillators.append(oscillator)
```

### Production code below!

```python
for inhib in range(len(edges)+1):
    counter= 0
    
    num_Os=32 # how many oscillators
    oscillators = []
    oscillator_config=[]
    # now manually create the desired component values below
    Rin_as=[*list(range(40,48)), *list(np.arange(40.5,48.5,1)), *list(np.arange(40.25,48.25,1)), *list(np.arange(40.375,48.375,1))]
    Ca_gnds= 32*[20]
    input_connections = [f'V{x}' for x in range(1,num_Vs+1)]
    output_connections = [dict() for i in range(num_Os)]

    for edge in edges:
        if counter >= inhib:
            output_connections[edge[0]][str(edge[1]+1)]= ['excitatory',{'R_coupling':'47k', 'C_coupling': '1u'}]
        
        else:
            output_connections[edge[0]][str(edge[1]+1)]=  ['inhibitory',{'R_coupling':'1'}]
            
        counter+=1
            

    #output_connections[0]= {'2': ['excitatory',{'R_coupling':'94k', 'C_coupling': '2u'}]}
    #output_connections[1] = {'3': ['inhibitory',{'R_coupling':'1'}]}
    for i in range(num_Os):
        oscillator_config.append({'Rin_a': f'{Rin_as[i]}k', 'Ca_gnd': f'{Ca_gnds[i]}u', 'input_connection': input_connections[i], 'output_connection': output_connections[i]})
        
    for i in range(num_Os):
        oscillator= create_oscillator(**oscillator_config[i])
        oscillators.append(oscillator)
    #if inhib==3 or inhib==96:
     #   print(oscillators)
    O_config = {'num_Os': num_Os, 'values': oscillators}
    circuit_config ={'V': V_config, 'O': O_config}
    
    circuit_name= f'inhibit_excitat_32Osci_{inhib}'
    with open(f'../data/inhibit_excit/32Oscis/{circuit_name}.dictionary', 'wb') as config_dictionary_file:

      pickle.dump(circuit_config, config_dictionary_file)
    
    netlist=create.Fireflies(f'../data/inhibit_excit/32Oscis/{circuit_name}.dictionary')
    netlist.create_netlist()
    netlist_file=netlist.return_netlist_file()
    with open(f'../data/inhibit_excit/32Oscis/{circuit_name}.txt', 'wb') as file:
        file.write(netlist_file)
    
```
