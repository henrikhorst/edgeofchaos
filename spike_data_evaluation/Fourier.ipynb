```python
#frequency boxes
freq_max=10
f=0.01
frames = np.linspace(0,freq_max,int(freq_max/f)+1)

#create empty lists to be filled with the values for the different distance measures 
frequency_counts=[]


#for each different coupling resistance 
for coupling_r in coupling_rs:
    path = sl.PATH + FOLDER + 'fireflys100_' + coupling_r + 'OhmCouplingResistance' + sl.END
    data = sl.load_spike_data(path)
    #create list to be filled with spike_trains each one corresponding to one oscillator
    fireflies_spike_frequencies=[]
    
    for elem in data:

        spike_frequency = 1/np.diff(elem)
        
        spike_frequency= np.digitize(spike_frequency,frames)
        fireflies_spike_frequencies.append(list(spike_frequency))
        
        
    fireflies_spike_frequencies= [elem for innerList in fireflies_spike_frequencies for elem in innerList]
    frequency_counts.append(len(np.unique(np.array(fireflies_spike_frequencies))))
    
```

```python


SAMPLE_RATE = 44100  # Hertz
DURATION = 5  # Seconds

def generate_sine_wave(freq, sample_rate, duration):
    x = np.linspace(0, duration, sample_rate * duration, endpoint=False)
    frequencies = x * freq
    # 2pi because np.sin takes radians
    y = np.sin((2 * np.pi) * frequencies)
    return x, y

# Generate a 2 hertz sine wave that lasts for 5 seconds
x, y = generate_sine_wave(200, SAMPLE_RATE, DURATION)
plt.plot(x, y)
```

```python
from scipy.fft import fft, fftfreq

# Number of samples in normalized_tone
N = SAMPLE_RATE * DURATION

yf = fft(y)
xf = fftfreq(N, 1 / SAMPLE_RATE)

plt.plot(xf, np.abs(yf))
```
